## 1.1 Dataset Source and Description

### Dataset Origin
The Bank Marketing Dataset is from the UCI Machine Learning Repository, created by:
- **Authors**: Sérgio Moro (ISCTE-IUL), Paulo Cortez (Univ. Minho), Paulo Rita (ISCTE-IUL)
- **Institution**: Portuguese Banking Institution
- **Time Period**: May 2008 to November 2010
- **Data Collection**: Direct marketing campaigns via phone calls

### Dataset Versions
**Original Dataset (bank-full.csv)**:
- **Instances**: 45,211
- **Features**: 16 input features + 1 target
- **Focus**: Basic client and campaign information

**Enhanced Dataset (bank-additional-full.csv)**:
- **Instances**: 41,188
- **Features**: 20 input features + 1 target
- **Additional Features**: 5 macroeconomic indicators from Banco de Portugal
  - emp.var.rate: employment variation rate (quarterly)
  - cons.price.idx: consumer price index (monthly)
  - cons.conf.idx: consumer confidence index (monthly)
  - euribor3m: euribor 3 month rate (daily)
  - nr.employed: number of employees (quarterly)

### Citations
1. **Original Dataset**: Moro et al. (2011) - "Using Data Mining for Bank Direct Marketing: An Application of the CRISP-DM Methodology"
2. **Enhanced Dataset**: Moro et al. (2014) - "A Data-Driven Approach to Predict the Success of Bank Telemarketing" - Decision Support Systems

## 1.4 Dataset Merging Strategy

### Challenge
- **bank-full.csv**: 45,211 rows × 16 features (missing 5 economic features)
- **bank-additional-full.csv**: 41,188 rows × 20 features (has all features)
- Different column structures require alignment before concatenation

### Solution Approach
1. **Load both datasets separately**
2. **Align columns**: Add missing economic features to bank-full.csv with NaN values
3. **Handle column differences**:
   - bank-full.csv has 'day' (day of month)
   - bank-additional-full.csv has 'day_of_week'
4. **Concatenate rows**: Stack datasets vertically
5. **Final dataset**: ~86,399 rows × 21 columns (20 features + target)

### Benefits
- Larger training dataset improves model generalization
- More diverse examples across different time periods
- Economic features available for subset of data (can be imputed if needed)

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')

print("Libraries imported successfully!")
```

```python
# Load the two datasets
print("Loading datasets...")
bank_full = pd.read_csv('../dataset/bank/bank-full.csv', sep=';')
bank_additional = pd.read_csv('../dataset/bank-additional/bank-additional-full.csv', sep=';')

print(f"\nbank-full.csv shape: {bank_full.shape}")
print(f"bank-additional-full.csv shape: {bank_additional.shape}")

print("\n" + "="*80)
print("bank-full.csv columns:")
print(bank_full.columns.tolist())

print("\n" + "="*80)
print("bank-additional-full.csv columns:")
print(bank_additional.columns.tolist())
```

```python
# Display first few rows of each dataset
print("Sample from bank-full.csv:")
display(bank_full.head())

print("\nSample from bank-additional-full.csv:")
display(bank_additional.head())
```

```python
# Identify column differences
bank_full_cols = set(bank_full.columns)
bank_additional_cols = set(bank_additional.columns)

print("Columns only in bank-full.csv:")
print(bank_full_cols - bank_additional_cols)

print("\nColumns only in bank-additional-full.csv:")
print(bank_additional_cols - bank_full_cols)

print("\nCommon columns:")
print(bank_full_cols & bank_additional_cols)
```

## 1.5 Dataset Alignment and Merging

```python
# Create a copy of bank_full and add missing economic features with NaN
bank_full_aligned = bank_full.copy()

# Add the 5 missing economic features with NaN
economic_features = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
for feature in economic_features:
    bank_full_aligned[feature] = np.nan

# Add day_of_week column to bank_full_aligned (will be NaN since we don't have this info)
bank_full_aligned['day_of_week'] = np.nan

# bank-additional has 'day_of_week' but not 'day'
# Add 'day' column to bank_additional for consistency
bank_additional_aligned = bank_additional.copy()
bank_additional_aligned['day'] = np.nan

print(f"bank_full_aligned shape: {bank_full_aligned.shape}")
print(f"bank_additional_aligned shape: {bank_additional_aligned.shape}")

print("\nColumns in bank_full_aligned:")
print(sorted(bank_full_aligned.columns.tolist()))

print("\nColumns in bank_additional_aligned:")
print(sorted(bank_additional_aligned.columns.tolist()))
```

```python
# Ensure both datasets have the same columns in the same order
all_columns = sorted(set(bank_full_aligned.columns) | set(bank_additional_aligned.columns))

# Reorder columns for both datasets
bank_full_aligned = bank_full_aligned.reindex(columns=all_columns)
bank_additional_aligned = bank_additional_aligned.reindex(columns=all_columns)

print("Column alignment complete!")
print(f"\nbank_full_aligned: {bank_full_aligned.shape}")
print(f"bank_additional_aligned: {bank_additional_aligned.shape}")

# Verify columns match
assert list(bank_full_aligned.columns) == list(bank_additional_aligned.columns), "Columns don't match!"
print("\n✓ Column structure verified!")
```

```python
# Add source column to track which dataset each row came from
bank_full_aligned['data_source'] = 'bank_full'
bank_additional_aligned['data_source'] = 'bank_additional'

# Concatenate the datasets
merged_data = pd.concat([bank_full_aligned, bank_additional_aligned], axis=0, ignore_index=True)

print(f"Merged dataset shape: {merged_data.shape}")
print(f"Expected rows: {len(bank_full) + len(bank_additional)} = {len(bank_full)} + {len(bank_additional)}")
print(f"Actual rows: {len(merged_data)}")
print(f"\n✓ Merging successful! Total records: {len(merged_data):,}")
```

```python
# Display merged dataset summary
print("Merged Dataset Summary:")
print("=" * 80)
print(merged_data.info())

print("\n" + "=" * 80)
print("Data source distribution:")
print(merged_data['data_source'].value_counts())

print("\n" + "=" * 80)
print("Target variable distribution:")
print(merged_data['y'].value_counts())
print(f"\nClass imbalance ratio: {merged_data['y'].value_counts()['no'] / merged_data['y'].value_counts()['yes']:.2f}:1")
```

```python
# Check missing values
print("Missing Values Analysis:")
print("=" * 80)
missing_summary = merged_data.isnull().sum()
missing_summary = missing_summary[missing_summary > 0].sort_values(ascending=False)
missing_pct = (missing_summary / len(merged_data) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing_summary,
    'Percentage': missing_pct
})

print(missing_df)

print("\nNote: Economic features and day_of_week have ~52% missing values")
print("These are from bank-full.csv which didn't have these features originally.")
```

```python
# Visualize class distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Overall class distribution
merged_data['y'].value_counts().plot(kind='bar', ax=axes[0], color=['#ff7f0e', '#2ca02c'])
axes[0].set_title('Overall Target Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Subscribed to Term Deposit', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)

# Add value labels on bars
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%d')

# Class distribution by data source
source_target = pd.crosstab(merged_data['data_source'], merged_data['y'])
source_target.plot(kind='bar', ax=axes[1], color=['#ff7f0e', '#2ca02c'])
axes[1].set_title('Target Distribution by Data Source', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Data Source', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
axes[1].legend(title='Subscribed', labels=['No', 'Yes'])

plt.tight_layout()
plt.savefig('../reports/figures/01_target_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Visualization saved to reports/figures/01_target_distribution.png")
```
